```python
import pandas as pd
```

```python
df=pd.read_excel('test_data (1).xlsx')
```

```python
df.head()
```

```python
df2=df.drop(columns=['ID'],axis=1)
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2['Hours']=df2['Date Time'].dt.hour
```

```python
df2.head()
```

```python
df2=df2.sort_values("Date Time")
```

```python
df2.head()
```

```python
df2=df2.set_index('Date Time')
```

```python
df2.head(15)
```

```python
sum_cols=['HAone','HAtwo','HS','HL','HD','HUO','HUC','AC-1 Current','AC-2 Current','Lobby Current',
    'Signage','DVR','UPS Output','UPS Current',
    'UPS Charging','UPS Discharging']
mean_cols = [
    'Temperature','Actual Temperature','Average Temperature',
    'Humidity','Raw Voltage','Earthing',
    'UPS ','Battery Voltage','ups_phase_to_earth_voltage',
    'PR','DR'
]

agg_dict={col:'sum' for col in sum_cols}
```

```python
agg_dict
```

```python
agg_dict.update({col:'mean' for col in mean_cols})
```

```python
agg_dict
```

```python
daily_df=df2.resample("D").agg(agg_dict)
print(daily_df)
```

```python
import numpy as np

days = np.arange(1, len(daily_df)+1).reshape(-1, 1)
```

```python
print(days)
daily_df
```

```python
from sklearn.linear_model import LinearRegression
day4_predictions={}

for col in daily_df.columns:
    y=daily_df[col].values

    model=LinearRegression()
    model.fit(days,y)

    day4_predictions[col]=model.predict([[len(days)+1]])[0]

day4_df=pd.DataFrame.from_dict(
    day4_predictions,orient='index',columns=['Day 4 Prediction']
)
day4_df
    
```

```python
future_preds={}

for col in daily_df.columns:
    y=daily_df[col].values

    model=LinearRegression()
    model.fit(days,y)


    preds=model.predict([[len(days)+2],[len(days)+3]])
    future_preds[col]=preds

    future_df = pd.DataFrame(
    future_preds,
    index=["Day 5 (Normal)", "Day 6 (Normal)"]
).T

future_df
```

```python
# Scneario prediction
# I have taken 2 scenarios where in the first there is a usage of upto 15% which is considered to be high and second of -10% which is low
scenario_df = future_df.copy()

scenario_df["Day 5 (High Load)"] = scenario_df["Day 5 (Normal)"] * 1.15
scenario_df["Day 6 (High Load)"] = scenario_df["Day 6 (Normal)"] * 1.15

scenario_df["Day 5 (Energy Saving)"] = scenario_df["Day 5 (Normal)"] * 0.90
scenario_df["Day 6 (Energy Saving)"] = scenario_df["Day 6 (Normal)"] * 0.90

scenario_df

```

```python
with pd.ExcelWriter("power_forecast_output.xlsx") as writer:
    daily_df.to_excel(writer, sheet_name="Daily_Aggregated_Data")
    day4_df.to_excel(writer, sheet_name="Day4_Prediction")
    scenario_df.to_excel(writer, sheet_name="Scenarios")

print("Forecast saved successfully")
```
